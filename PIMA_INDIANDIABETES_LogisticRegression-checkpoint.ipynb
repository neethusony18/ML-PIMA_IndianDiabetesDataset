```python
#import the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the dataset
df=pd.read_csv("C:\\Users\\subin\\OneDrive\\Documents\\Neethu\\Data Science\\pimaindiansdiabetescsv\\pima-indians-diabetes.csv")
df.head()
#giving column names to the dataset
df.columns=["Preg","Glucose","BP","Skin_Thick","Insulin","BMI","Diabetes_PedFun","Age","Class"]
df.head()
```

```python
#data wrangling
df.isnull().sum()
#out put shows its a clean data
#data visualization
df.hist(bins=50,color="blue",figsize=(20,15))
plt.show()
```

```python
df.head(20)
#output shows that there are many values for the coloumns that are zero.
median_BP=df['BP'].median()
df['BP']=df['BP'].replace(to_replace=0,value=median_BP)
median_Skin_Thick=df['Skin_Thick'].median()
df['Skin_Thick']=df['Skin_Thick'].replace(to_replace=0,value=median_Skin_Thick)
median_Insulin=df['Insulin'].median()
df['Insulin']=df['Insulin'].replace(to_replace=0,value=median_Insulin)
median_BMI=df['BMI'].median()
df['BMI']=df['BMI'].replace(to_replace=0,value=median_BMI)
median_Glucose=df['Glucose'].median()
df['Glucose']=df['Glucose'].replace(to_replace=0,value=median_Glucose)
median_Age=df['Age'].median()
df['Age']=df['Age'].replace(to_replace=0,value=median_Age)
df.head(20)
```

```python
#We require the columns from 0 to 7 excluding the class column
X=df.iloc[:,0:7]
Y=df["Class"]
```

```python
#Split the dataset into the training and the test data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=9)
#print(X_train.head())
#print(Y_train.head())
#print(X_test.head())
print(Y_test.head())
```

```python
#Building the model
from sklearn.linear_model import LogisticRegression
reg_model=LogisticRegression().fit(X_train,Y_train)
predictions=reg_model.predict(X_train)
print("The predictions are:",predictions[0:5])
```

```python
#Evaluation on the training test
score=reg_model.score(X_test,Y_test) 
print("The accuracy of the predicted model with the test data is:",score)
```
